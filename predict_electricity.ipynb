```python
timezonedf = {'America/New_York', 'Europe/London', 'America/Phoenix', 'America/Chicago', 'America/Los_Angeles', 'Europe/Zurich', 'Asia/Singapore', 'America/Denver'}
industrydf = {'Education', 'Government', 'Commercial Property'}
```

```python
while True:
    timezone = str(input("Enter timezone: "))
    if timezone not in timezonedf:
        print("Invalid Answer")
        continue
    else:
        break
```

```python
while True:
    industry=str(input("Please enter the industry "))
    if industry not in industrydf:
        print("This is an invalid answer")
        continue
    else:
        break
```
